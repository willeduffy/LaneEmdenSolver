```python
#Defining constants

k = 1.38064852e-23 #J/k
h = 6.626070040e-34 ##Js
mh = 1.660549e-27 #kg
mue = 1.17
mu = 1.29 + 1.17
emass = 9.10938356e-31 #kg
c = 299792458 #m/s

#making an array of temperature values from 3 billion to 6 billion
# counting by 10^7, to be plotted as independent variable
t = np.arange(3e9,6e9,1e7)
```

#### Classical ideal gas pressure vs. Electron degeneracy gas pressure   
$k_1 (\dfrac{\rho}{\mu_e})^\frac{5}{3} = \dfrac{\rho kT}{\mu m_H}$

$\rho = (\dfrac{kT\mu_e^\frac{5}{3}}{\mu m_H k_1})^\frac{3}{2}$

$k_1^\frac{3}{2} = \dfrac{3}{\pi}\dfrac{h^3}{20^\frac{3}{2}m_e^\frac{3}{2}m_H^\frac{5}{2}}$

$\dfrac{m_H^\frac{5}{2}}{m_H^\frac{3}{2}} = m_H$

$\rho = \dfrac{40\sqrt{5}\pi \mu_e^\frac{5}{2}k^\frac{3}{2}m_e^\frac{3}{2}m_H}{3h^3\mu^\frac{3}{2}} * T^\frac{3}{2}$

```python
# Classical vs. electron degeneracy pressure slope

p = 40 * 5**.5 * mue**(5/2) * np.pi * mh * (k*emass*t/mu)**(3/2) / (3* h**3)
```

#### Classical ideal gas pressure vs. Relativistic electron degeneracy pressure
$k_2 (\dfrac{\rho}{\mu_e})^\frac{4}{3} = \dfrac{\rho kT}{\mu m_H}$

$\rho = (\dfrac{kT\mu_e^\frac{4}{3}}{\mu m_H k_2})^3 = \dfrac{k^3T^3\mu_e^4}{\mu^3 m_H^3 k_2^3}$

$k_2^3 = \frac{3}{\pi}\dfrac{h^3c^3}{8^3m_H^4}$

$\dfrac{m_H^4}{m_H ^3} = m_H$

$\rho = \dfrac{512\mu_e^4 \pi m_H k^3}{3c^3h^3\mu^3}*T^3$

```python
# Classical vs. relativistic electron degeneracy pressure slope

p2 = (512 * mue**4 * np.pi * k**3 * mh * t**3) / (3 * c**3 * h**3 * mu**3)
```

#### Electron degeneracy pressure vs. Relativistic electron degeneracy pressure
$k_1(\dfrac{\rho}{\mu_e})^\frac{5}{3} = k_2(\dfrac{\rho}{\mu_e})^\frac{4}{3}$

$\rho = \dfrac{k_2^3}{k_1^3}\mu_e$

$k_1^3 = \dfrac{9}{\pi^3}\dfrac{h^6}{20^3m_e^3m_H^5}$

$\dfrac{k_2^3}{k_1^3} \rightarrow \dfrac{3}{\pi}*\dfrac{\pi^2}{9} = \dfrac{\pi}{3}$

$\dfrac{h^3c^320^3m_e^3}{8^3h^6} = \dfrac{125c^3m_e^3}{8h^3}$

$\dfrac{m_H^5}{m_H^4} = m_H$

$\rho = \dfrac{125 c^3 me_e^3\mu_e \pi m_H}{24h^3}$

```python
# degeneracy pressure vs. relativistic degeneracy pressure

p3 = (125 * mue * np.pi * c**3 * mh * emass**3)/(24*h**3)

# need an array of the same value of p3 so that numpy can 
# iterate over it to make a graph - p3 is temperature independent
p3array = []

for i in t:
    p3array.append(p3)
```

### 3. 

Lane-Emden Equation: $\dfrac{1}{\xi^2}\dfrac{d}{d\xi}(\xi^2\dfrac{d\theta}{d\xi}) = -\theta^n$   

Boundary Condition 1: at $\xi = 0, \theta = 1$   
Boundary Condition 2: at $\xi = 0, \dfrac{d\theta}{d\xi} = 0$

### n = 0  
$\dfrac{1}{\xi^2}\dfrac{d}{d\xi}(\xi^2\dfrac{d\theta}{d\xi}) = -1$   

$\dfrac{d}{d\xi}(\xi^2\dfrac{d\theta}{d\xi}) = -\xi^2$

$\int d(\xi^2\dfrac{d\theta}{d\xi}) = \int -\xi^2 d\xi$

$\xi^2\dfrac{d\theta}{d\xi} = -\frac{1}{3}\xi^3 + C_1$

By invoking Boundary Condition 2, $\dfrac{d\theta}{d\xi} = -\dfrac{1}{3}\xi + \dfrac{C_1}{\xi^2} \rightarrow 0 = \dfrac{C_1}{0} \rightarrow C_1 = 0$

$\dfrac{d\theta}{d\xi} = -\dfrac{1}{3}\xi$

$\int d\theta = \int -\dfrac{1}{3}d\xi$

$\theta = -\dfrac{1}{6}\xi^2 + C_2$

By invoking Boundary Condition 1, $\theta = 1 = C_2$

### $\theta_0 = 1 - \dfrac{1}{6}\xi^2$

### n = 1   

$\dfrac{1}{\xi}\dfrac{d}{d\xi}(\xi^2\dfrac{d\theta}{d\xi}) = -\theta$

Substitute $\chi = \xi\theta \rightarrow \theta = \dfrac{\chi}{\xi}$

$\dfrac{d\theta}{d\xi} \rightarrow$ Quotient rule $\rightarrow \dfrac{1}{\xi}\dfrac{d\chi}{d\xi} - \dfrac{\chi}{\xi^2}$

Substituting that expression into the original Lane-Emden equation for $\dfrac{d\theta}{d\xi}$ yields $\dfrac{1}{\xi^2}\dfrac{d}{d\xi}(\xi\dfrac{d\chi}{d\xi}-\chi) = \dfrac{-\chi}{\xi}$

$\dfrac{d}{d\xi}[\xi\dfrac{d\chi}{d\xi} - \chi] = \xi\dfrac{d^2\chi}{d\xi^2} + \dfrac{d\chi}{d\xi} - \dfrac{d\chi}{d\xi} = \xi\dfrac{d^2\chi}{d\xi^2}$

$\dfrac{1}{\xi}\dfrac{d^2\chi}{d\xi^2} = \dfrac{-\chi}{\xi}$

$\dfrac{d^2\chi}{d\xi^2} = -\chi$ which is indicative of simple harmonic motion where $k = 1$

The general solution for simple harmonic motion is $\chi(\xi) = Asin(\xi) + Bcos(\xi)$

To solve for the constants, first invoke Boundary Condition 1. At $\xi = 0$,  

$\chi = \xi * \theta = 0 * 1 = 0$

$0 = Asin(0) + Bcos(0) \rightarrow B = 0$

$\chi(\xi) = Asin(\xi)$

To solve for A, first get the equation in terms of $\theta$ again by dividing by $\xi$   

$\theta(\xi) = A\dfrac{sin(\xi)}{\xi}$

To find A, evaluate the limit as $\xi \rightarrow 0$

$\lim_{\xi\to0} \dfrac{sin(\xi)}{\xi} = \dfrac{0}{0} \rightarrow$ L'Hpital's rule

$\lim_{\xi\to0} \dfrac{cos(\xi)}{1} = 1 \rightarrow A = 1$

### $\theta_1 = \dfrac{sin\xi}{\xi}$

#### Numerical solutions for any value of n

In order to use the fourth-order Runge-Kutta method, we must get the Lane-Emden equation in the form of    
$\frac{dy}{dt} = f(t,y)$   
$y(t_0) = y_0$   

Here it is assumed that $t_0$ and $y_0$ are known. The rate of change of y depends on t and y itself.

For easier coding, $\xi$ will be referred to as $x$ and $\theta$ will be reffered to as $y$. Thus,

Lane-Emden Equation: $\dfrac{1}{x^2}\dfrac{d}{dx}(x^2\dfrac{dy}{dx}) = -y^n$  



$x^2 \dfrac{d^2y}{dx^2} + 2x\dfrac{dy}{dx} + x^2y^n = 0$

Now to get it into the proper form, make the substitution $z = \frac{dy}{dx}$

$\dfrac{dy}{dx} = z$   
$\dfrac{dz}{dx} = -\dfrac{1}{x^2}(2xz + x^2y^n)$

```python
# define functions used in integrate function
# corresponds to dy/dx
def f(x, y, z):
    return z

# corresponds to dz/dx
def g(x, y, z, n):
    return -(2*x*z + x**2 * y**n)/x**2

# integrates for a single step
# x_0, y_0, z_0 are initial values for each variable
# n is corresponding n in Lane-Emden Equation
# h is step size
def integrate(x_0, y_0, z_0, n, h):
    k_0 = h * f(x_0, y_0, z_0)
    l_0 = h * g(x_0, y_0, z_0, n)
    k_1 = h * f(x_0+1/2*h, y_0+1/2*k_0, z_0+1/2*l_0)
    l_1 = h * g(x_0+1/2*h, y_0+1/2*k_0, z_0+1/2*l_0, n)
    k_2 = h * f(x_0+1/2*h, y_0+1/2*k_1, z_0+1/2*l_1)
    l_2 = h * g(x_0+1/2*h, y_0+1/2*k_1, z_0+1/2*l_1, n)
    k_3 = h * f(x_0+h, y_0+k_2, z_0+l_2)
    l_3 = h * g(x_0+h, y_0+k_2, z_0+l_2, n)
    
    x_1 = x_0 + h
    y_1 = y_0 + 1/6 * (k_0+2*k_1+2*k_2+k_3)
    z_1 = z_0 + 1/6 * (l_0+2*l_1+2*l_2+l_3)
    
    return (x_1, y_1, z_1)
```

```python
#grapher returns a list of x and y values, the length of which 
# depends on how long it takes for y to go negative

def grapher(x_0, y_0, z_0, n, h):
    #xs and xy start as lists with only the initial values in them, but they will be added to
    xs = [x_0]
    ys = [y_0]
    while y_0>=0:
        #change x_0, y_0, and z_0 to the values output by the integrate function
        x_0, y_0, z_0 = integrate(x_0, y_0, z_0, n, h)
        
        #then add them to their respective list
        xs.append(x_0)
        ys.append(y_0)
    
    #returns a list of lists
    return [xs, ys]

#define special grapher for n = 5 case in order to display it nicely -- for displaying purposes don't let x get bigger than 16
def special_grapher(x_0, y_0, z_0, n, h):
    xs = [x_0]
    ys = [y_0]
    #special case where x stops before 16 for graphing purposes
    while y_0>=0 and x_0 < 16:
        x_0, y_0, z_0 = integrate(x_0, y_0, z_0, n, h)
        xs.append(x_0)
        ys.append(y_0)
        
    #returns a list of lists
    return [xs, ys]
```

```python
fig3=plt.figure(1,figsize=(18,5))  
ax3=fig3.add_subplot(111)          
    
#display the solutions for n = 0 to n = 5
for i in range(6):
    if i<5:
        ax3.plot(grapher(1e-50, 1, 0, i, .1)[0], grapher(1e-50, 1, 0, i, .1)[1], label="n = "+str(i))
    #use special_grapher for n = 5
    if i == 5:
        ax3.plot(special_grapher(1e-50, 1, 0, i, .1)[0], special_grapher(1e-50, 1, 0, i, .1)[1], label="n = 5")
    
        
ax3.set_title('Lane-Emden equation for varying n', fontsize=18)
ax3.set_xlabel("", fontsize=18)
ax3.set_ylabel('', fontsize=18)

# Place a legend to the right.
ax3.legend(loc=1, borderaxespad=0.5)
```
